## Предобработка данных

Приведем названия всех столбцов к змеиному стилю:

```python
data = data.rename(columns={'cityCenters_nearest': 'city_centers_nearest', 'studio': 'is_studio', 'open_plan': 'is_open_plan'})
data.columns
```

### Обработка пропусков

Изучим сколько пропусков в данных:

```python
data.isna().sum()
```

Также посмотрим на долю пропусков в данных:

```python
data.isna().mean()
```

Имеются пропуски в таких столбац как: ceiling_height, floors_total, living_area, is_apartment, kitchen_area, balcony, locality_name, airports_nearest, city_centers_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition.

Продумаем для каждого столбца на что изменять пропуски.
- для столбца ceiling_height пропуски заменим на среднее значение по столбцу
- для столбца floors_total заменим пропуски на среднее значение по каждому городу
- для столбца living_area сложно подобрать значение на которое будем менять пропуск, поскольку на площадь квартиры влияет много параметров, а от нее уже зависит цена, поэтому пропуски оставим
- для столбца is_apartment проанализируем взаимосвязи с столбцами is_studio и is_open_plan
- для столбца kitchen_area также сложно определить на что менять пропуск как и со столбцом living_area
- для столбца balcony пропуски заменим на 0, поскольку если человек не указал количество балконов, то скорее всего их нет
- для столбца locality_name пропуски заменим на 'неизвестно'
- для столбцов airports_nearest, city_centers_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest пропуски будем менять на средние значения по городу
- для столбца days_exposition пропуски в данных могут означать что объявление зполнили в день выгрузки данных (необходимо предварительно посмотреть на уникальные значения)

Замена пропусков в столбце ceiling_height на среднее значение по столбцу:

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].mean())
data['ceiling_height'].isna().sum()
```

Замена пропусков в столбцах: floors_total, airports_nearest, city_centers_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest на среднее значение по каждому городу для каждого параметра:

```python
city_list = list(data['locality_name'].unique())
parametr_list = ['floors_total', 'airports_nearest', 'city_centers_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest']

for city in city_list:
    for parametr in parametr_list:
        parametr_mean = data[data['locality_name'] == city][parametr].mean()
        data.loc[data['locality_name'] == city, parametr] = data.loc[data['locality_name'] == city, parametr].fillna(parametr_mean)
    
data.isna().sum()
```

Нашлось лишь одно объявление в котором невозможно определить среднюю величину по количеству этажей относительно города, поэтому просто заменим этот пропуск на среднее количестов этажей по всему датасету:

```python
data[data['floors_total'].isna()]
```

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].mean())
data['floors_total'].isna().sum()
```

Заменим пропуски в столбце balcony:

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].isna().sum()
```

Далее проработаем столбец days_exposition

Напишем функцию, которая пройдется по трем столбцам, и если в одном из трех уже стоит True, то в остальных двух столбах точно должны быть False.

```python
data.head()
```

Проверим есть ли вообще строки в которых одновременно два из трех вариантов равны True:

```python
data[(data['is_studio']==True) & (data['is_open_plan']==True)]
```

Поскольку таких строк в таблице не оказалось, а в стобцах is_studio и is_open_plan нет пропусков, то пройдемся по каждому из них, и если встретим True, то заполним в столбце is_apartment значение False, а если в обоих столбцах is_studio и is_open_plan будет False, то в is_apartment должно быть True.

```python
def apartment_check(row):
    if row['is_studio'] == True:
        row['is_apartment'] = False
    if row['is_open_plan'] == True:
        row['is_apartment'] = False
    if row['is_studio'] == False and row['is_open_plan'] == False:
        row['is_apartment'] = True
    return row

data = data.apply(apartment_check, axis=1)
data.head()
```

### Изучите предложения квартир: для каждой квартиры есть информация о расстоянии до центра. Выделите квартиры в Санкт-Петербурге (locality_name). Ваша задача — выяснить, какая область входит в центр. Создайте столбец с расстоянием до центра в километрах: округлите до целых значений. После этого посчитайте среднюю цену для каждого километра. Постройте график: он должен показывать, как цена зависит от удалённости от центра. Определите границу, где график сильно меняется, — это и будет центральная зона.

Добавим столбец с расстоянием до центра в километрах.

```python
def div1000(distance):
    result = round(distance / 1000)
    return result

data['city_centers_nearest_km'] = data['city_centers_nearest'].apply(div1000)
```

Сделаем срез по данным Санкт-Петербурга:

```python
data_spb = data.query('locality_name == "Санкт-Петербург"')
data_spb.head()
```

Найдем среднюю цену для каждого километра:

```python
data_spb_grouped = data_spb.groupby('city_centers_nearest_km')['last_price'].mean()
data_spb_grouped
```

Построим график зависимости цены от удаленности от центра:

```python
data_spb_grouped.plot(grid=True, linewidth=3, legend=True, title='price_per_km')
```

**Вывод**

Для всех объявлений Санкт-Петербурга мы рассчитали среднюю цену в зависимости от удаленности от центра города. График показывает что происходит резкое изменение цены в районе 7-8 километров от центра.
