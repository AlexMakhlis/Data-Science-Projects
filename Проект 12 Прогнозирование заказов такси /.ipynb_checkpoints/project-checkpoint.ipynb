#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Подготовка

Импортируем необходимые библиотеки и загрузим данные:

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
from catboost import CatBoostRegressor

import lightgbm as lgb
from lightgbm import LGBMRegressor

import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('/datasets/taxi.csv', index_col=[0], parse_dates=[0])
data.head()
```

Отсортируем данные по хронологии:

```python
data.sort_index(inplace=True)
data.index.is_monotonic
```

```python
data.info()
```

Выполним ресемплирование данных по часу:

```python
data = data.resample('1H').sum()
data.head()
```

**Вывод:**

Мы произвели загрузку данных, преобразовали столбец с датами к типу datetime, провели ресемплирование по одному часу. Явные проблемы в данных отсутствуют, поэтому можно переходить к анализу данных и дальнейшему моделлированию.

## Анализ

Взглянем на данные в целом:

```python
data.plot()
plt.title('График зависимости заказов такси от времени')
plt.show()
```

Найдем скользящее среднее:

```python
data.rolling(24).mean().plot(figsize=(10,5))
plt.title('Скользящее среднее')
plt.show()
```

Мы добавили скользящее среднее с окном равным одни сутки. Из анализа за пол года можно судить о том, что к лету в целом увеличился вызов такси. Проверим еще сезонность и тренд:

```python
def seasonality_and_trend(data):
    decomposed = seasonal_decompose(data) 

    plt.figure(figsize=(10, 12))

    plt.subplot(311)
    decomposed.trend.plot(ax=plt.gca())
    plt.title('Тренд')

    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca())
    plt.title('Сезонность')

    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca())
    plt.title('Шум')
    plt.tight_layout()
    
seasonality_and_trend(data)
```

Посмотрим теперь на один месяц:

```python
seasonality_and_trend(data['2018-03-01':'2018-04-01'])
```

Возьмем данные за одну неделю:

```python
seasonality_and_trend(data['2018-03-16':'2018-03-23'])
```

Взглянем еще на один случайный день:

```python
seasonality_and_trend(data['2018-03-20':'2018-03-21'])
```

**Вывод**

Можно утверждать по графику сезонности за день, что есть нагрузка на заказ такси в 12:00 и в 18:00. По поводу сезонности месячной нельзя утвеждать, поскольку данных для этого анализа мало (половинга года). А вот по графику тренда за неделю можно предположить, что количество заказов к выходным растет.

## Обучение

Добавим признаков в данные для обучения моделей:

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

make_features(data, 4, 6)
data.head()
```

```python
data = data.dropna()
```

```python
data.head()
```

```python
features = data.drop('num_orders', axis=1)
target = data['num_orders']
print(features.shape)
print(target.shape)
```

```python
features_main, features_test, target_main, target_test = train_test_split(features, target, test_size=0.1, shuffle=False)
features_train, features_valid, target_train, target_valid = train_test_split(features_main, target_main, test_size=0.25, shuffle=False)
del features_main
del target_main
display(features_train.shape)
display(target_train.shape)
display(features_valid.shape)
display(target_valid.shape)
display(features_test.shape)
display(target_test.shape)
```

### Линейная регрессия

```python
model = LinearRegression()
model.fit(features_train, target_train)
```

```python
predictions_train = model.predict(features_train)
rmse = mean_squared_error(target_train, predictions_train) ** 0.5
print(f'RMSE_train = {rmse}')
```

```python
predictions_valid = model.predict(features_valid)
rmse = mean_squared_error(target_valid, predictions_valid) ** 0.5
print(f'RMSE_valid = {rmse}')
```

```python
columns = ['linear', 'CatBoost', 'LGBM', 'Decision_Tree', 'Random_Forest']
indexes = ['RMSE']
table_rmse = pd.DataFrame(columns=columns, index=indexes)
table_rmse['linear']['RMSE'] = rmse
table_rmse
```

```python
predictions_valid_series = pd.Series(predictions_valid, index=target_valid.index)
table_predict = pd.DataFrame(target_valid)
table_predict['predict_linear'] = predictions_valid_series
table_predict.plot(figsize=(10,6))
plt.title('Сравнение предсказаний')
plt.show()
```

**Вывод**

По графику видно, что линейная регрессия не особо улавливает всплески, которые как раз нам и нужны.

### CatBoost

```python
model = CatBoostRegressor(loss_function="RMSE")

parameters = {'max_depth' : [4, 6, 8],
              'learning_rate' : [0.2, 0.4, 0.6],
              'iterations' : [100, 200, 300, 500, 1000]
                 }
tscv = TimeSeriesSplit()
grid = GridSearchCV(model, parameters, cv=tscv)
grid.fit(features_train, target_train, verbose=10)
```

```python
grid.best_params_
```

```python
model = CatBoostRegressor(loss_function="RMSE", iterations=100, max_depth=6, learning_rate=0.2)
model.fit(features_train, target_train, verbose=10)
```

```python
predictions_valid = model.predict(features_valid)
rmse = mean_squared_error(target_valid, predictions_valid) ** 0.5
print(f'RMSE_valid = {rmse}')
```

```python
table_rmse['CatBoost']['RMSE'] = rmse
table_rmse
```

```python
predictions_valid_series = pd.Series(predictions_valid, index=target_valid.index)
table_predict['predict CatBoost'] = predictions_valid_series
table_predict.plot(y =['num_orders','predict CatBoost'], figsize=(10,6))
plt.title('Сравнение предсказаний')
plt.show()
```

### LightGBM

```python
model = LGBMRegressor()

parameters = {'max_depth' : [2, 4, 6, 8],
              'learning_rate' : [0.1, 0.2, 0.4, 0.6],
              'n_estimators' : [100, 200, 300, 500]
                 }
tscv = TimeSeriesSplit()
grid = GridSearchCV(model, parameters, cv=tscv)
grid.fit(features_train, target_train, eval_metric='rmse')
```

```python
grid.best_params_
```

```python
model = LGBMRegressor(n_estimators=100, max_depth=8, learning_rate=0.1)
model.fit(features_train, target_train, eval_metric='rmse')
predictions_valid = model.predict(features_valid)
rmse = mean_squared_error(target_valid, predictions_valid) ** 0.5
print(f'RMSE_valid = {rmse}')
```

```python
table_rmse['LGBM']['RMSE'] = rmse
table_rmse
```

```python
predictions_valid_series = pd.Series(predictions_valid, index=target_valid.index)
table_predict['predict LGBM'] = predictions_valid_series
table_predict.plot(y =['num_orders','predict LGBM'], figsize=(10,6))
plt.title('Сравнение предсказаний')
plt.show()
```

### Решающее дерево

```python
model = DecisionTreeRegressor(random_state=27)
parameters = {'max_depth' : [2, 3, 4, 5, 6, 8, 10],
              'min_samples_split' : [2, 3, 4, 5, 6],
              'min_samples_leaf' : [2, 3, 4, 5, 6]
                 }
tscv = TimeSeriesSplit()
grid = GridSearchCV(model, parameters, cv=tscv, scoring=mean_squared_error)
grid.fit(features_train, target_train)
```

## Тестирование

Для предсказаний мы использовали такие модели как: линейная регрессия, CatBoost, LGBM, Решающее дерево и Случайный лес. Нашей задачей был поиск модели и лучших гиперпараметров для предсказания с наименьшим RMSE. Результаты исследования представлены ниже:

```python
table_rmse
```

Из полученных показателей видно, что лучшее значение RMSE показала модель LGBM.

Протестируем теперь лучшую модель и получим итоговое значение RMSE. Обучать модель будем сразу на train и valid.

```python
features = data.drop('num_orders', axis=1)
target = data['num_orders']
print(features.shape)
print(target.shape)
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.1, shuffle=False)

display(features_train.shape)
display(target_train.shape)

display(features_test.shape)
display(target_test.shape)
```

```python
model = LGBMRegressor(n_estimators=100, max_depth=8, learning_rate=0.1)
model.fit(features_train, target_train, eval_metric='rmse')
predictions_test = model.predict(features_test)
rmse = mean_squared_error(target_test, predictions_test) ** 0.5
print(f'RMSE_test = {rmse}')
```

```python
predictions_test_series = pd.Series(predictions_test, index=target_test.index)
table_predict_test = pd.DataFrame(target_test)
table_predict_test['predict_LGBM'] = predictions_test_series
table_predict_test.plot(figsize=(10,6))
plt.title('Сравнение предсказаний')
plt.show()
```

Из графика видно, что модель в целом неплохо отрабатывает всплески.

Также проверим модель на адекватность. Сравним модель с моделью, предсказывающей значение на основе предыдущего:

```python
pred_previous = target_test.shift()
pred_previous = pred_previous.fillna(target_train.iloc[-1])
print("RMSE:",  (mean_squared_error(target_test, pred_previous) ** 0.5))
```

Таким образом, модель можно считать адекватной.
